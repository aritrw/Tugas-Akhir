Library yang akan digunakan

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

Load Dataset

```python
stunting_dataset = pd.read_csv('Dataset_Skripsi.csv')
```

```python
stunting_dataset.head()
```

```python
stunting_dataset.info()
```

```python
from IPython.display import display
display(stunting_dataset)
```

```python
stunting_dataset.shape 
```

```python
duplikat_kolom = stunting_dataset[
    stunting_dataset.duplicated(
        subset=["Usia", "Jenis_Kelamin", "BB_Lahir", "TB_Lahir",
                "BB_Sekarang", "TB_Sekarang", "Stunting"]
    )
]

if duplikat_kolom.empty:
    print("Tidak ada data duplikat berdasarkan kolom yang dipilih.")
else:
    print("Data duplikat ditemukan:")
    print(duplikat_kolom)
```

Pengecekan Missing Value

```python
stunting_dataset.isnull().sum()
```

Label Encoding

```python
stunting_dataset['Jenis_Kelamin'] = stunting_dataset['Jenis_Kelamin'].replace({'Laki-laki':1,'Perempuan':0})
stunting_dataset['Stunting'] = stunting_dataset['Stunting'].replace({'No':0,'Yes':1})
```

```python
# memisahkan data dan label
X = stunting_dataset.drop (columns='Stunting', axis=1)
Y = stunting_dataset['Stunting'] 
```

```python
print(X) 
```

```python
print(Y)
```

Standarisasi Data

```python
scaler = StandardScaler()
scaler.fit(X)
standarized_data = scaler.transform(X) 
X = standarized_data
Y = stunting_dataset['Stunting'] 
print(X)
print(Y)
```

Memisahkan Data Training dan Data Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, stratify=Y, random_state=42)
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV

# Inisialisasi model SVM
svm_model = SVC(probability=True, random_state=42)

# Parameter grid untuk dicoba
param_grid = {
    'C': [0.1, 1, 10, 100],             # regularisasi
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],  # jenis kernel
}

grid_search = GridSearchCV(
    estimator=svm_model,
    param_grid=param_grid,
    cv=5,            # 5-fold cross validation
    n_jobs=-1,       # pakai semua core CPU
    verbose=2,       # tampilkan progres
    scoring='accuracy'  # metrik evaluasi
)

# Jalankan pencarian grid
grid_search.fit(X_train, Y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```

Membuat data latih menggunakan algoritma Random Forest


```python
model_svm = SVC(C=10, kernel='rbf')
model_svm.fit(X_train, Y_train)
```

Membuat model evaluasi untuk mengukur tingkat akurasi

```python
sv = model_svm.predict(X_train)
training_data_accuracy = accuracy_score(sv,Y_train)
print('Akurasi data training adalah =', training_data_accuracy)


sv = model_svm.predict(X_test)
test_data_accuracy = accuracy_score(sv,Y_test)
print('Akurasi data testing adalah =', test_data_accuracy)
```

```python
from sklearn.metrics import confusion_matrix

# Prediksi data uji
y_pred = model_svm.predict(X_test)

# Hitung confusion matrix
cm = confusion_matrix(Y_test, y_pred)
TN, FP, FN, TP = cm.ravel()

# Hitung metrik
sensitivitas = TP / (TP + FN)  
spesifisitas = TN / (TN + FP)  

print("Confusion Matrix:\n", cm)
print(f"Sensitivitas (Recall): {sensitivitas}")
print(f"Spesifisitas: {spesifisitas}")
```

Membuat model deteksi

```python
input_data = (7,1,3.8,49,28,72.2)

input_data_as_numpy_array = np.array(input_data)

input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshape)
print(std_data)

prediction = model_svm.predict(std_data)
print(prediction)

if (prediction[0] == 0):
    print('Anak tidak berisiko stunting')
else :
    print('Anak berisiko stunting') 
```
